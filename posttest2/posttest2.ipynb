### 3. Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.

```python
from prettytable import PrettyTable #mengimport prettytable, untuk membuat tampilan table lebih rapi

jumlah_record = sb.size
jumlah_atribut = len(sb.columns)
nama_atribut = sb.columns
jumlah_atribut = len(sb.select_dtypes('int').columns)

info = {
    "Jumlah Record" : jumlah_record,
    "Jumlah Attribute" : jumlah_atribut,
    "Nama Attribute" : nama_atribut,
    "Jumlah Attribute bertipe Angka" : jumlah_atribut
}

table = PrettyTable()
for i in info.keys():
    table.add_column(i, [])
table.add_row([info.get(i, "") for i in info.keys()])
print(table)
```

##### 5. Memvisualisasikan dataset dengan grafik histogram

```python
import matplotlib.pyplot as plt

sb['Year'].value_counts()

fig = plt.figure(figsize=(20,15))

plt.hist(sb['Year'], bins = 15, color = 'blue', edgecolor = 'white')
plt.title('Country League')
plt.ylabel('Match Played')
plt.xlabel('Year')
```

##### 6. Menampilkan korelasi dataset menggunakan heatmap.

### HeatMap
import library seaborn untuk memvisualisasikan HeatMap

```python
import seaborn as sns
```

```python
sns.heatmap(
    sb.corr(),
    annot=True,
    annot_kws={'size':5}
)

plt.show()
```
